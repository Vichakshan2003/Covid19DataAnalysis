```python
vaccine_df = pd.read_csv("covid_vaccine_statewise.csv")
```

```python
vaccine_df.head(7)
```

```python
# dropping Time, ConfirmedIndianations, ConfirmedForeignNational, and Sno number

covid_df.drop(["Sno", "Time", "ConfirmedIndianNational", "ConfirmedForeignNational"], inplace = True, axis = 1)
```

```python
covid_df.head()
```

```python
covid_df.dropna(subset=['Date'], inplace=True)
covid_df['Date']= pd.to_datetime(covid_df['Date'], format = '%Y-%m-%d')
```

```python
covid_df.head()
```

```python
# Active Cases = Total Confirmed Cases - (Cured + Death )

covid_df['Active_Cases'] = covid_df['Confirmed'] - (covid_df['Cured'] + covid_df['Deaths'])
covid_df.tail()
```

```python
statewise = pd.pivot_table(covid_df, values = ["Confirmed", "Deaths", "Cured"],
                           index = "State/UnionTerritory", aggfunc = max)
```

```python
# recovery rate
statewise["Recovery Rate"] = statewise["Cured"]*100/statewise["Confirmed"]
```

```python
# mortality rate
statewise["Mortality Rate"] = statewise["Deaths"]*100/statewise["Confirmed"]
```

```python
statewise = statewise.sort_values(by = "Confirmed", ascending = False)
```

```python
statewise.style.background_gradient(cmap= "cubehelix")
```

```python
# Tp 10 states based on Active Cases
top_10_active_cases = (covid_df.groupby(by = 'State/UnionTerritory').max()
                           [['Active_Cases', 'Date']].
                           sort_values(by =['Active_Cases'], ascending = False).reset_index())

fig = plt.figure(figsize = (16,9))
plt.title("Top 10 States with most active cases in India", size = 25)
ax = sns.barplot(data = top_10_active_cases.iloc[:10],
                 y = "Active_Cases", x = "State/UnionTerritory",
                 linewidth = 2, edgecolor = 'red')


plt.xlabel("States")
plt.ylabel("Total Active Cases")
plt.show()
```

```python
# Top 10 states with highest Deaths

top_10_deaths = (covid_df.groupby(by = 'State/UnionTerritory').max()
[['Deaths','Date']].sort_values(by = ['Deaths'], ascending = False).reset_index())

plt.figure(figsize = (18,5))
plt.title("Top 10 states with highest Deaths", size = 25)

ax = sns.barplot(data = top_10_deaths.iloc[:12],
                 y = "Deaths",
                 x = "State/UnionTerritory",
                 linewidth = 2,
                 edgecolor = "black")

plt.xlabel("States")
plt.ylabel("Total Deaths Cases")
plt.show()
```

```python
# growth trend

fig = plt.figure(figsize=(12,6))

ax = sns.lineplot(data = covid_df[covid_df['State/UnionTerritory'].isin(['Maharashtra', 'Karnataka', 'Kerala', 'Tamil Nadu','Uttar Pradesh'])],
                                  x ='Date', y = 'Active_Cases', hue = 'State/UnionTerritory')

ax.set_title("Top 5 effect states in India", size = 16)
```

```python
# dropping missign valyes columns

vaccination = vaccine_df.drop(columns = ['Sputnik V (Doses Administered)',
                                         'AEFI','18-44 Years (Doses Administered)',
                                         '45-60 Years (Doses Administered)',
                                         '60+ Years (Doses Administered)'],
                                          axis = 1)

vaccination.head()
```

```python
# male vs female vaccination (pie plot)

male = vaccination["Male(Individuals Vaccinated)"].sum()
female = vaccination["Female(Individuals Vaccinated)"].sum()

px.pie(names=["Male", "Female"], values=[male, female], title = "Male vs Female Vaccination")
```

```python
# Remove the rows where State = india

vaccine = vaccine_df[vaccine_df.State!='India']
vaccine
```

```python
vaccine.rename(columns = {"Total Individuals Vaccinated": "Total"}, inplace = True)
vaccine
```

```python
# most vaccinated state

max_vac = vaccine.groupby('State')['Total'].sum().to_frame('Total')
max_vac = max_vac.sort_values(by='Total',ascending=False)[:5]
max_vac
```

```python
fig = plt.figure(figsize=(10, 5))
plt.title("Top 5 Vaccinated States in India", size = 20)
x = sns.barplot(data = max_vac.iloc[:10], y = max_vac.Total, x= max_vac.index, linewidth = 2, edgecolor = "red")
plt.xlabel("States")
plt.ylabel("Vaccination")
plt.show()
```
